# Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

This task will be done in following steps:

1. First get the webpage https://www.shine.com/
2. Enter “Data Analyst” in “Job title, Skills” field and enter “Bangalore” in “enter the location” field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.

```python
import selenium
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings("ignore")
import time
import keyboard

from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
```

```python
driver = webdriver.Chrome()
driver.get("https://www.shine.com/")
```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException

try:
    wait = WebDriverWait(driver, 25)
    element = wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'modal_notification__rwztg')))
    notification = driver.find_element(By.CLASS_NAME, "btn-close")
    notification.click()
except TimeoutException:
    print("Element not found within 10 seconds")
```

```python
desig=driver.find_element(By.CLASS_NAME,"searchBarInput")   
desig.click()
```

```python
time.sleep(2)
desig=driver.find_element(By.CLASS_NAME,"form-control  ")
desig.send_keys("Data Analyst")
```

```python
#keyboard.press_and_release('tab')
time.sleep(2)
location=driver.find_element(By.XPATH,"/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input")
location.send_keys("Bangalore")
```

```python
time.sleep(2)
r_click=driver.find_element(By.CLASS_NAME,"styledcls_body_hidden")
r_click.click()
```

```python
time.sleep(2)
search=driver.find_element(By.CLASS_NAME,"searchForm_btnWrap_advance__VYBHN")
search.click()
```

```python
try:
    x=WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, "//button[@class='btn-close register_close']")))
    x.click()
except Exception as e:
    print(e)
```

```python
job_titles = []
locations = []
company_names = []
experiences = []

job_title = driver.find_elements(By.XPATH,'//h2[@itemprop="name"]/a')

for i in job_title:
    title = i.text.strip()
    job_titles.append(title)
    
location = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"][1]')

for l in location:
    title = l.text.replace("\n"," ")
    title = title.split()[0]
    locations.append(title)
    
    
company_name = driver.find_elements(By.XPATH,('//div[@class="jobCard_jobCard_cName__mYnow"]/span'))

for c in company_name:
    title = c.text
    company_names.append(title)
    
experience = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')

for e in experience:
    title = e.text
    experiences.append(title)
    
    
df=pd.DataFrame({"Job-titles":job_titles,"Job-Location":locations, "Company_name":company_names, "experience_required":experiences})
df.index +=1
df.head(10)
```
