# Q2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. 

Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. 

In case, if any of the details are missing for any of the product then replace it by “-“.

```python
Brand_Name = []
Name_of_Product=[]
Ratings=[]
Price = []
Return_Exchange = []
Expected_Delivery = []
Availability = []
Product_URL = []
```

```python
pages=[]
start=0
end=3

for page in range(start,end):
    page=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in page:
        pages.append(i.get_attribute('href'))

next_page=driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
next_page.click()
len(pages)
```

```python
for i in pages:
    driver.get(i)
    time.sleep(1)

    try:
        brand=driver.find_element(By.XPATH,'//tr[@class="a-spacing-small po-brand"]')
        Brand_Name.append(brand.text.split()[1])

    except:
        Brand_Name.append('-')
        
    try:
        NOP=driver.find_element(By.XPATH,'//span[@class="a-size-large product-title-word-break"]')
        Name_of_Product.append(NOP.text)

    except:
        Name_of_Product.append('-')
    
    try:
        rating=driver.find_element(By.XPATH,"//span[@class='a-size-base a-nowrap']")
        Ratings.append(rating.text.split()[0])

    except:
        Ratings.append('-')
        
          
    try:
        price=driver.find_element(By.XPATH,'//*[@id="corePrice_feature_div"]/div/div/span[1]/span[2]/span[2]')
        Price.append(price.text)

    except:
        Price.append('-')
     
    try:
        exchange=driver.find_element(By.XPATH,'//*[@id="RETURNS_POLICY"]/span/div[2]/span')
        Return_Exchange.append(exchange.text)

    except:
        Return_Exchange.append('-')    
    
        
    try:
        Delivery=driver.find_element(By.XPATH,'//*[@id="mir-layout-DELIVERY_BLOCK-slot-PRIMARY_DELIVERY_MESSAGE_LARGE"]/span/span')
        Expected_Delivery.append(Delivery.text)

    except:
        Expected_Delivery.append('-')
          
                    
    try:
        avail=driver.find_element(By.XPATH,'//*[@id="availability"]/span')
        Availability.append(avail.text)

    except:
        Availability.append('-')
        
        
len(Brand_Name),len(Name_of_Product),len(Ratings),len(Price),len(Return_Exchange),len(Expected_Delivery),len(Availability),len(pages)
```

```python
len(Brand_Name),len(Name_of_Product),len(Ratings),len(Price),len(Return_Exchange),len(Expected_Delivery),len(Availability),len(pages)
```

```python
data=pd.DataFrame({'Brand Name':Brand_Name,'Name of Product':Name_of_Product,"Ratings":Ratings,"Price":Price,"Return/Exchange":Return_Exchange,"Expected Delivery":Expected_Delivery,"Availability":Availability,"Product URL":pages})
data.index +=1
data
```

```python
data.to_csv('Question_2.csv', index=True)
```

Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”,
“Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. 

Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
import selenium
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings("ignore")
import time
import keyboard

from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
```

```python
time.sleep(1)
search=driver.find_element(By.CLASS_NAME,'Pke_EE')
search.send_keys('Samsung Galaxy Note')
```

```python
searchbutton=driver.find_element(By.CLASS_NAME,'_2iLD__')
searchbutton.click()
```

```python
Brand_Name = []
Smartphone_name =[]
Colour = []
RAM = []
Storage_ROM = []
Primary_Camera = []
Secondary_Camera = []
Display_Size = []
Battery_Capacity = []
Price = []
Product_URL = []

brandname=driver.find_elements(By.XPATH,'//div[@class="_4rR01T"]')
for bn in brandname:
    try:
        Brand_Name.append(bn.text.split()[0])
    except:
        Brand_Name.append('-')
        
smartphone=driver.find_elements(By.XPATH,'//div[@class="_4rR01T"]')
for sp in smartphone:
    try:
        Smartphone_name.append(sp.text)
    except:
        Smartphone_name.append("-")
        
color=driver.find_elements(By.XPATH,'//div[@class="_4rR01T"]')
for c in color:
    try:
        Colour.append(c.text)
    except:
        Colour.append("-")
        
price=driver.find_elements(By.XPATH,'//div[@class="_30jeq3 _1_WHN1"]')
for p in price:
    try:
        Price.append(p.text)
    except:
        price.append("-")
        
urls = driver.find_elements(By.XPATH,'//a[@class="_1fQZEK"]')
for u in urls:
    try:
        Product_URL.append(u.get_attribute("href"))
    except:
        Product_URL.append("-")
```

```python
df1 = pd.DataFrame({'Brand_Name':Brand_Name,'Smartphone_name': Smartphone_name,'Colour':Colour,'Price':Price})

df1['Colour'] = df1['Colour'].str.extract(r'\((.*?)\,')[0].tolist()

df5=pd.DataFrame({'Product_URL':Product_URL})
```

```python
tags=driver.find_elements(By.XPATH,'//ul[@class="_1xgFaf"]')
table=[]

for li in tags:
    item=driver.find_elements(By.XPATH,'//li')
    row=[item.text for item in li.find_elements(By.XPATH,'.//li' )]
    table.append(row)
```

```python
df2=pd.DataFrame(table)

df2.columns=['RAM|ROM|UPTO','Display_Size ','Primary_Camera|Secondary_Camera','Battery_capacity','Processor','Warranty']
df2.rename(columns={'Storage':0,'Camera':1,'Display':2,'Battery':3,'Processor':4,'Warranty':5,})

df3 = df2['Primary_Camera|Secondary_Camera'].str.split('|', expand=True)
df2['Primary Camera'] = df3[0]
df2['Secondary Camera'] = df3[1]

df2.drop('Primary_Camera|Secondary_Camera', axis=1, inplace=True)
df2.drop('Warranty', axis=1, inplace=True)
df2.drop('Processor', axis=1, inplace=True)

df4=pd.concat([df1,df2,df5],axis=1)
df4.index +=1
df4
```

```python
df4.to_csv("Question_4.csv",index=True)
```
