# 3) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape and make data frame-

a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
page = requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")

soup = BeautifulSoup(page.content, "html.parser")
table1 = soup.find("table", class_="table")

data = []
for row in table1.tbody.find_all("tr"):
    x = row.find_all("td")
    x = [col.text.replace("\n"," ").strip() for col in x]
    
    data.append(x)

df = pd.DataFrame(data, columns=["Position", "Team", "Matches", "Points", "Rating"])
df.set_index("Position", inplace=True)
top_10_ODI_teams = df.head(10)

top_10_ODI_teams
```

# 5) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame-
i) Headline
ii) Time
iii) News Link

```python
import datetime

url = 'https://www.cnbc.com/world/?region=world'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

headlines = []
times = []
news_links = []

for article in soup.find_all('div', class_="LatestNews-headlineWrapper"):
    headline = article.find('a').text.strip()
    time = article.find('time').text.strip()
    news_link = article.find('a')['href']
    headlines.append(headline)
    times.append(time)
    news_links.append(news_link)

df = pd.DataFrame({'Headline': headlines, 'Time': times, 'News Link': news_links})
df.index +=1
df
```

# 6) Write a python program to scrape the details of most downloaded articles from AI in last 90 days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles Scrape below mentioned details and make data frame-
i) Paper Title
ii) Authors
iii) Published Date
iv) Paper URL

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd


url = 'https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

titles = []
authors = []
dates = []
urls = []

for article in soup.find_all('div', class_="sc-orwwe2-3 jOMrrY"):
    title = article.find_all("h2", class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg")
    title_text = [t.text.strip() for t in title]
    author = article.find_all('span', class_='sc-1w3fpd7-0 dnCnAO')
    author_text = [a.text.strip() for a in author]
    date = article.find_all('span', class_='sc-1thf9ly-2 dvggWt')
    date_text = [d.text.strip() for d in date]
    url = article.find_all('a')
    url_text = [u['href'] for u in url]
    titles.extend(title_text)
    authors.extend(author_text)
    dates.extend(date_text)
    urls.extend(url_text)

df6 = pd.DataFrame({'Paper Title': titles, 'Authors': authors, 'Published Date': dates, 'Paper URL': urls})
df6.index +=1
df6.index.name = 'Article_Position'
df6
```

# 7) Write a python program to scrape mentioned details from dineout.co.in and make data frame   
i)Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
url = "https://www.dineout.co.in/delhi-restaurants/buffet-special"
page = requests.get(url)
soup = BeautifulSoup(page.content, "html.parser")

resta_name1= []
resta_cuisines2= []
resta_location3= []
rating4= []
image5= []

for x in soup.find("div", class_="restnt-card-wrap-new"):
    name=x.find_all("a", class_="restnt-name ellipsis")
    resta_name=[t.text.strip() for t in name]
    resta_name1.extend(resta_name)
    
    cuisine=x.find_all("span",class_="double-line-ellipsis")
    resta_cuisines=[t.text.strip() for t in cuisine]
    cuisines = [g.split('|')[1].strip() for g in resta_cuisines]
    resta_cuisines2.extend(cuisines)
        
    location=x.find_all("div", class_="restnt-loc ellipsis")
    resta_location=[t.text.strip() for t in location]
    resta_location3.extend(resta_location)
    
    rating =x.find_all("div", class_="restnt-rating rating-4")
    resta_rating=[t.text.strip() for t in rating]
    rating4.append(resta_rating)
    
for i in soup.find_all("img", class_="no-img"):
    image5.append(i["data-src"])

    
df7=pd.DataFrame({"Restaurant Name": resta_name1,"Cuisine": resta_cuisines2,
                  "Location": resta_location3,"Ratings": rating4,"Image URL": image5,})

df7.index +=1
df7.index.name = 'Dine-in@Delhi'
df7
```
