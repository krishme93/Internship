# 1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: 

A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
import pandas as pd
import selenium
import time
import keyboard
import requests
import regex as re
import warnings
warnings.filterwarnings("ignore")

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.common.exceptions import NoSuchElementException
```

```python
driver=webdriver.Chrome()
time.sleep(1)

driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
time.sleep(1)
```

```python
Rank =[]
Name =[]
Artist =[]
Upload_date=[]
Views=[]

try:
    rank_tags=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")
    for i in rank_tags[0:30]:
        Rank.append(i.text.replace(".",""))
except NoSuchElementException:
    Rank.append('-')
except StaleElementReferenceException:
    Rank.append('-')    

try:
    name_tags=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")
    for i in name_tags[0:30]:
        Name.append(i.text)
except NoSuchElementException:
    Name.append('-')
except StaleElementReferenceException:
    Name.append('-')
    
try:
    artist=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
    for i in artist[0:30]:
        Artist.append(i.text)
except NoSuchElementExceptionhElementException:
    Artist.append('-')
except StaleElementReferenceException:
    Artist.append('-')
    
try:
    upload_date=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")
    for i in upload_date[0:30]:
        Upload_date.append(i.text)
except NoSuchElementExceptionhElementException:
    Upload_date.append('NA')
except StaleElementReferenceException:
    Upload_date.append('NA')

    
try:
    views=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")
    for i in views[0:30]:
        Views.append(i.text)
except NoSuchElementExceptionhElementException:
    Views.append('NA')
except StaleElementReferenceException:
    Views.append('NA')
    
df=pd.DataFrame({'Rank':Rank,'Video Name':Name,'Uploader':Artist,'Views (in Billons)':Views,'Upload Date':Upload_date})
#df.index +=1 
#df.index.name = "Rank"
df.name = "Most viewed videos on YouTube from Wikipedia"
df1 = df.style.set_table_styles([dict(selector='th', props=[('text-align', 'center')])])
df1.set_properties(**{'text-align': 'center'}).hide_index()

df1
```

```python
driver.close()
```

# 5. Scrape the details of top 100 songs on billiboard.com. 

Url = https:/www.billboard.com/ You have to find the following details:

A) Song name
B) Artist name
C) Last week rank
D) Peak rank
E) Weeks on board

Note: - From the home page you have to click on the charts option then hot 100-page link through code.

```python
driver=webdriver.Chrome()
url='https://www.billboard.com/charts/hot-100/'
driver.get(url)
time.sleep(1)
driver.maximize_window()
```

```python
Song =[]
Artist =[]
Last_Week_rank=[]
Peak_rank =[]
Weeks =[]


try:
    song=driver.find_elements(By.XPATH,'//li[@class="lrv-u-width-100p"]/ul/li/h3')
    for i in song:
        Song.append(i.text)
except NoSuchElementException:
    Song.append('NA')
    
# Scraping Song Artist Name
try:
    artist=driver.find_elements(By.XPATH,'//li[@class="lrv-u-width-100p"]/ul/li[1]/span[1]')
    for i in artist:
        Artist.append(i.text)
except NoSuchElementException:
    Artist.append('NA')
    
# Scraping Song last week rank
try:
    last_rank=driver.find_elements(By.XPATH,'//li[@class="o-chart-results-list__item // a-chart-color u-width-72 u-width-55@mobile-max u-width-55@tablet-only lrv-u-flex lrv-u-flex-shrink-0 lrv-u-align-items-center lrv-u-justify-content-center lrv-u-border-b-1 u-border-b-0@mobile-max lrv-u-border-color-grey-light u-background-color-white-064@mobile-max u-hidden@mobile-max"][1]')
    for i in last_rank:
        Last_Week_rank.append(i.text)
except NoSuchElementException:
    Last_Week_rank.append('NA')

# Scraping Song peak rank
try:
    peak=driver.find_elements(By.XPATH,'//li[@class="o-chart-results-list__item // a-chart-bg-color a-chart-color u-width-72 u-width-55@mobile-max u-width-55@tablet-only lrv-u-flex lrv-u-flex-shrink-0 lrv-u-align-items-center lrv-u-justify-content-center lrv-u-background-color-grey-lightest lrv-u-border-b-1 u-border-b-0@mobile-max lrv-u-border-color-grey-light u-hidden@mobile-max"][2]')
    for i in peak:
        Peak_rank.append(i.text)
except NoSuchElementException:
    Peak_rank.append('NA')

# Scraping Song weeks on chart
try:
    weeks=driver.find_elements(By.XPATH,'//li[@class="o-chart-results-list__item // a-chart-color u-width-72 u-width-55@mobile-max u-width-55@tablet-only lrv-u-flex lrv-u-flex-shrink-0 lrv-u-align-items-center lrv-u-justify-content-center lrv-u-border-b-1 u-border-b-0@mobile-max lrv-u-border-color-grey-light u-background-color-white-064@mobile-max u-hidden@mobile-max"][2]')
    for i in weeks:
        Weeks.append(i.text)
except NoSuchElementException:
    Weeks.append('NA')

                              
                              
df=pd.DataFrame({'Song_Name':Song,'Artist_Name':Artist,'Last_week_rank':Last_Week_rank,'Peak':Peak_rank,'Weeks_on_chart':Weeks})
df.index +=1
df.index.name="Rank"
df
```

B) Year span
C) Genre

```python
driver=webdriver.Chrome()
url='https://www.imdb.com/list/ls095964455/'
driver.get(url)
time.sleep(1)
driver.maximize_window()
```

```python
Name =[]
Year_Span=[]
Genre =[]
Run_Time =[]
Ratings =[]
Votes =[]

try:
    name=driver.find_elements(By.XPATH,'//h3[@class="lister-item-header"]/a')
    for i in name:
        Name.append(i.text)
except NoSuchElementException:
    Name.append('NA')

try:
    year=driver.find_elements(By.XPATH,'//span[@class="lister-item-year text-muted unbold"]')
    for i in year:
        Year_Span.append(i.text)
except NoSuchElementException:
    Year_Span.append('NA')
    

try:
    genre=driver.find_elements(By.XPATH,'//span[@class="genre"]')
    for i in genre:
        Genre.append(i.text)
except NoSuchElementException:
    Genre.append('NA')
    

try:
    runtime=driver.find_elements(By.XPATH,'//span[@class="runtime"]')
    for i in runtime:
        Run_Time.append(i.text)
except NoSuchElementException:
    Run_Time.append('NA')


try:
    ratings=driver.find_elements(By.XPATH,'//div[@class="ipl-rating-star small"]/span[2]')
    for i in ratings:
        Ratings.append(i.text)
except NoSuchElementException:
    Ratings.append('NA')


try:
    votes=driver.find_elements(By.XPATH,'//div[@class="lister-item-content"]/p[4]/span[2]')
    for i in votes:
        Votes.append(i.text)
except NoSuchElementException:
    Votes.append('NA')
    
df7=pd.DataFrame({"Name":Name,"Year Span":Year_Span,"Genre":Genre,"Run Time":Run_Time,"Ratings":Ratings,"Votes":Votes})
df7.index.name = "Most watched Tv series-rank"
df7.index += 1
df7
```

```python
driver=webdriver.Chrome()
url='https://archive.ics.uci.edu/datasets'
driver.get(url)
time.sleep(1)
driver.maximize_window()
```

```python
try:
    click=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/div/div[2]/button')
    click.click()
    
except:
    print('no cookies available')
```

```python
pages=[]
start=0
end=65

for page in range(start,end):
    page=driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
    for i in page:
        pages.append(i.get_attribute('href'))
    next_page=driver.find_element(By.XPATH,'//button[@class="btn-primary btn-sm btn"][2]')
    next_page.click()
    time.sleep(2)
len(pages)
```

```python
Dataset =[]
Data_Type =[]
Task =[]
Attribute_Type =[]
No_of_Instances =[]
No_of_Attribute =[]
Year =[]

for i in pages:
    driver.get(i)
    time.sleep(2)

    try:
        dataset=driver.find_element(By.XPATH,'//h1[@class="text-3xl font-semibold text-primary-content"]')
        Dataset.append(dataset.text)
    except:
        Dataset.append('-')
        
    
    try:
        datatype=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[1]/p')
        Data_Type.append(datatype.text)
    except:
        Data_Type.append('-')
        

    try:
        task=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[3]/p')
        Task.append(task.text)
    except:
        Task.append('-')
        
        
    try:
        attributetype=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[4]/p')
        Attribute_Type.append(attributetype.text)
    except:
        Attribute_Type.append('-')
        

        
    try:
        no_of_instances=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[5]/p')
        No_of_Instances.append(no_of_instances.text)
    except:
        No_of_Instances.append('-')
        
        
    try:
        no_of_attribute=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[6]/p')
        No_of_Attribute.append(no_of_attribute.text)
    except:
        No_of_Attribute.append('-')
        
        

    try:
        year=driver.find_element(By.XPATH,'//h2[@class="text-sm text-primary-content"]')
        Year.append(year.text.split()[2])
    except:
        Year.append('-')
        
        
        
        
len(Dataset),len(Data_Type),len(Task),len(Attribute_Type),len(No_of_Instances),len(No_of_Attribute),len(Year)
```

```python
df8=pd.DataFrame({'DataSet Title':Dataset,'Data Type':Data_Type,'Task':Task,'Attribute Type':Attribute_Type,
                'No of Instances':No_of_Instances,
                'No of Attribute':No_of_Attribute,
                'Year':Year})

df8.index +=1
df8.index.name ='UCI-ML repositories Datasets'
df8
```
